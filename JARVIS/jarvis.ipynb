```python
import pyttsx3
import speech_recognition as sr
import webbrowser
import time
import math
import requests
import re
import music_lib

# Initialize TTS and recognizer
engine = pyttsx3.init()
engine.setProperty('rate', 200)
r = sr.Recognizer()
r.energy_threshold = 4000

# Speak
def speak(text):
    print("Jarvis:", text)
    engine.say(text)
    engine.runAndWait()

# Prime check
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Table printer
def print_table(n):
    for i in range(1, 11):
        speak(f"{n} x {i} = {n * i}")

# Math solver
def solve_math(cmd):
    try:
        if "table of" in cmd:
            number = int(cmd.split("of")[1].strip())
            speak(f"Table of {number}")
            print_table(number)

        elif "square root of" in cmd:
            number = int(cmd.split("of")[1].strip())
            result = number ** 0.5
            speak(f"Square root of {number} is {result}")

        elif "divide" in cmd and "by" in cmd:
            words = cmd.split()
            index = words.index("by")
            num1 = int(words[index - 1])
            num2 = int(words[index + 1])
            result = num1 / num2
            speak(f"{num1} divided by {num2} is {result}")

        elif "to the power" in cmd:
            base = int(cmd.split("to the power")[0].strip())
            exponent = int(cmd.split("to the power")[1].strip())
            result = base ** exponent
            speak(f"{base} to the power {exponent} is {result}")

        elif "factorial of" in cmd:
            number = int(cmd.split("of")[1].strip())
            result = math.factorial(number)
            speak(f"Factorial of {number} is {result}")

        elif "percent of" in cmd:
            percent = float(cmd.split("percent of")[0].strip())
            total = float(cmd.split("percent of")[1].strip())
            result = (percent / 100) * total
            speak(f"{percent} percent of {total} is {result}")

        elif "is" in cmd and "prime" in cmd:
            number = int(cmd.split("is")[1].split("prime")[0].strip())
            if is_prime(number):
                speak(f"{number} is a prime number.")
            else:
                speak(f"{number} is not a prime number.")

        elif "even" in cmd or "odd" in cmd:
            number = int([word for word in cmd.split() if word.isdigit()][0])
            if number % 2 == 0:
                speak(f"{number} is even.")
            else:
                speak(f"{number} is odd.")

        elif "area of circle with radius" in cmd:
            radius = float(cmd.split("radius")[1].strip())
            area = math.pi * radius ** 2
            speak(f"Area of circle with radius {radius} is {area:.2f}")

        else:
            # Fallback to eval for expressions like 5 + 2 * 3
            result = eval(cmd)
            speak(f"The answer is {result}")

    except Exception as e:
        print("Sorry, I couldn't understand that math.")
        print("Error:", e)

def news():
    print("In news function")
    api_key="23a56b7c2264b60b37243b7d223e5ee4"
    country="in"
    r=requests.get(f"https://gnews.io/api/v4/top-headlines?country={country}&lang=en&max=5&apikey={api_key}")
    if r.status_code==200:
        data=r.json()
        articles=data.get('articles',[])
        for article in articles:
            speak(article['title'])

def songs(cmd):
    if cmd.lower()=="stealth":
        song=cmd.lower()
        link=music_lib.music[song]
        webbrowser.open(link)
    elif cmd.lower()=="sunny":
        song=cmd.lower()
        link=music_lib.music[song]
        webbrowser.open(link)
    elif cmd.lower()=="island":
        song=cmd.lower()
        link=music_lib.music[song]
        webbrowser.open(link)
    else:
        speak("Not Found")

# Process command
def process(cmd):
    cmd = cmd.lower().strip()

    if "open google" in cmd:
            webbrowser.open("https://google.com")
    elif "open youtube" in cmd:
            webbrowser.open("https://youtube.com")
    elif "open facebook" in cmd:
            webbrowser.open("https://facebook.com")
    elif "open instagram" in cmd:
            webbrowser.open("https://instagram.com")
    elif "open linkedin" in cmd:
            webbrowser.open("https://linkedin.com")
    else:
            speak("Sorry, I didn't recognize that command.")

# Startup guide
def show_commands():
    print("\nüéôÔ∏è Say 'Jarvis', then try:")
    print("   ‚Üí 'open google', 'open youtube'")
    print("   ‚Üí '5 + 10', 'table of 7'")
    print("   ‚Üí 'square root of 81', 'factorial of 6'")
    print("   ‚Üí 'area of circle with radius 5'")
    print("   ‚Üí 'is 19 prime', 'is 20 even'\n")

# Main loop
if __name__ == "__main__":
    show_commands()
    speak("Jarvis is now online.")
    speak("Activate Me By saying JARVIS")
    speak("Select A Mode")
    speak("News")
    speak("Songs")
    speak("Command")
    speak("Calculation")

    while True:
        try:
            with sr.Microphone() as source:
                print("üéß Listening for wake word...")
                r.adjust_for_ambient_noise(source, duration=1)
                audio = r.listen(source, timeout=3, phrase_time_limit=3)
                word = r.recognize_google(audio)

                # If speach recognition not wroks than manualy inputs
                # word=input("enter the name 1 :")

            if "jarvis" in word.lower():
                print("‚úÖ Wake word detected.")
                speak("I'm listening...")

                with sr.Microphone() as source: 
                    r.adjust_for_ambient_noise(source, duration=1)
                    audio = r.listen(source, timeout=5, phrase_time_limit=5)
                    command = r.recognize_google(audio)
                    print(command)
                    # If speach recognition not wroks than manualy inputs
                    # command=input("Enter the command")

                print(f"üó£Ô∏è You said: {command}")
                if "news" in command.lower():
                    speak("News Mode ON")
                    news()
                elif "songs" in command.lower():
                        speak("Songs Mode On")
                        with sr.Microphone() as source:
                            print("üéß Listening for wake word...")
                            r.adjust_for_ambient_noise(source, duration=1)
                            audio = r.listen(source, timeout=3, phrase_time_limit=3)
                            cmd = r.recognize_google(audio)
                            # cmd=input("Enter the song name")
                            songs(cmd)
                elif "command" in command.lower():
                    speak("Command mode on")
                    with sr.Microphone() as source:
                        print("üéß Listening for wake word...")
                        r.adjust_for_ambient_noise(source, duration=1)
                        audio = r.listen(source, timeout=3, phrase_time_limit=3)
                        command = r.recognize_google(audio)
                        # command=input("Enter the song name")
                        process(command)
                elif "calculation" in command.lower():
                    speak("Calculation mode on")
                    with sr.Microphone() as source:
                        print("üéß Listening for wake word...")
                        r.adjust_for_ambient_noise(source, duration=1)
                        audio = r.listen(source, timeout=3, phrase_time_limit=3)
                        command = r.recognize_google(audio)
                        # command=input("Enter the Expression")
                        solve_math(command)
                else:
                    speak("Invalid Command")

        except sr.WaitTimeoutError:
            print("‚è±Ô∏è Timeout")
        except sr.UnknownValueError:
            print("‚ùå Could not understand audio")
        except KeyboardInterrupt:
            print("üõë Exiting by user")
            break
        except Exception as e:
            print("‚ö†Ô∏è Error:", e)
```
