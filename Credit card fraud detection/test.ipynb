```python
import pandas as pd
import pickle

with open('ccfd_model.pkl','rb') as f:
    model=pickle.load(f)
with open('scaler.pkl','rb') as f:
    scaler=pickle.load(f)

df=pd.read_csv(r"C:\Users\Lenovo\Downloads\archive (10)\creditcard.csv")

if 'Class' in df.columns:
    df=df.drop(columns=['Class'])

sample=df.sample(5)
scaled=scaler.transform(sample)
predictions=model.predict(scaled)

for i in range(len(predictions)):
    print(f"\n Transaction {i+1}")
    print(sample.iloc[i])
    print("Prediction :","Fraud Detected" if predictions[i]==1 else "Safe")
```
